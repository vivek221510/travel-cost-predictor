```python
import pandas as pd

df=pd.read_csv("final_transport_effect_dataset.csv")
df["total_cost"] = df["total_cost"] * 0.5

df["cost_bin"] = pd.qcut(df["total_cost"], q=5, labels=False)
```

```python
X=df.drop(["total_cost","cost_bin"],axis=1)
y=df["total_cost"]

X=X.drop(["travel_cost","hotel_cost","food_cost"],axis=1)

X=pd.get_dummies(X,drop_first=True)      #makes text values to seperate columns and make them boolean.
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,y,test_size=0.2,stratify=df["cost_bin"],random_state=42
)
```

```python
from sklearn.ensemble import RandomForestRegressor

model=RandomForestRegressor(n_estimators=300,max_depth=15,min_samples_split=5,min_samples_leaf=2,random_state=42,n_jobs=-1)

model.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

pred_values=model.predict(X_test)

mae=mean_absolute_error(y_test,pred_values)
rmse=np.sqrt(mean_squared_error(y_test,pred_values))

print("MAE:",mae)
print("RMSE",rmse)
```
